```python
import numpy as np
from IPython.display import display_html
```

```python
def display_side_by_side(*args):
    html_str = ''
    for df in args:
        html_str += df.to_html()
    display_html(
        html_str.replace('table','table style="display:inline"'), 
        raw=True
    )
```

```python
weatherData["weather_code"][0]
```

```python
# weatherData on left, bestMatchData on right
display_side_by_side(weatherData[weatherData['weather_code'] != bestMatchData['weather_code']][['datetime','weather_code']], 
                     bestMatchData[weatherData['weather_code'] != bestMatchData['weather_code']][['datetime','weather_code']])
```

The two fields are different, even though in many cases weather_code on right is weather_code on left minus 1...

```python
display_side_by_side(weatherData[weatherData['direct_normal_irradiance_instant'] != bestMatchData['direct_normal_irradiance_instant']][['datetime','direct_normal_irradiance_instant']], 
                     bestMatchData[weatherData['direct_normal_irradiance_instant'] != bestMatchData['direct_normal_irradiance_instant']][['datetime','direct_normal_irradiance_instant']])
```

Above we have pretty significant difference and no apparent relations
