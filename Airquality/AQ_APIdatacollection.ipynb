```python
import pandas as pd 
import requests
```

```python
url = "https://air-quality-api.open-meteo.com/v1/air-quality?latitude=45.4643&longitude=9.1895&current=european_aqi&hourly=pm10,pm2_5,nitrogen_dioxide,sulphur_dioxide,ozone&start_date=2019-12-31&end_date=2023-12-31&domains=cams_europe"
```

```python
response = requests.get(url)
```

```python
response_json = response.json()
response_json
```

```python
aq_df = pd.DataFrame(response_json.get('hourly',[]))
display(aq_df)
```

```python
aq_df["time"] = pd.to_datetime(aq_df["time"])
```
