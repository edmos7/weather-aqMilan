#### Let's re-assess data quality of our daily air quality dataset after aggregation

Reminder: we want our data to be accurate(at least syntactically), complete and consistent

#### Syntactic Accuracy

Logically if values didn't exceed reference domain aggregations shouldn't either, but just to be sure:

```python
maxes = {
    'nitrogen_dioxide': 1000,
    'ozone': 800,
    'sulphur_dioxide': 1250,
    'pm2_5': 800,
    'pm10': 1200
}
```

```python
def checkRangeOfReference_max_agg(pollutant,maximum,agg,data):
    print("For "+pollutant+" :")
    for agg in aggregations:
        col = agg+pollutant
        n = data[data[col] > maximum].shape[0]
        print(col+" :"+str(n))
```

```python
for pollutant in pollutants:
    checkRangeOfReference_max_agg(pollutant,maxes[pollutant],aggregations,daily_aq)
```

```python
def checkRangeOfReference_min_agg(pollutant,minimum,agg,data):
    print("For "+pollutant+" :")
    for agg in aggregations:
        col = agg+pollutant
        n = data[data[col] < minimum].shape[0]
        print(col+" :"+str(n))
```

```python
for pollutant in pollutants:
    checkRangeOfReference_min_agg(pollutant,0,aggregations,daily_aq)
```

```python
import re
```

```python
# Next let's make sure min/max_time columns are in a reasonable range (from 00 to 23)
def check_time_format(data,col,pollutant):
    # Define the regex pattern (values from 00 to 23, followed by :00:00)
    pattern = r"^(?:[01]\d|2[0-3]):00:00$"

    # Apply regex check to the column
    def validate_format(value):
        return bool(re.match(pattern, value))

    invalid = pd.DataFrame()
    invalid["is_valid"] = data[col+pollutant].apply(validate_format)
    display(invalid)

    # Get the entries that do not follow the format
    invalid_entries = invalid[~invalid["is_valid"]]

    # Print the number of invalid entries and their indices
    num_invalid = len(invalid_entries)
    print(f"Number of invalid entries for {col} in {pollutant}: {num_invalid}")
```

```python
time_cols = ['min_time_','max_time_']
```

```python
for pollutant in pollutants:
    for col in time_cols:
        check_time_format(daily_aq,col,pollutant)
```
